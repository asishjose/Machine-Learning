```python
from skimage import io
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
img = io.imread('bird_small.png')

pixels = img.reshape(-1,3)
```

```python
pixels
```

```python
img.shape
```

```python
pixels.shape
```

```python
kmeans = KMeans(n_clusters=8, random_state=0).fit(pixels)
```

```python
compressed = kmeans.cluster_centers_[kmeans.labels_]
compressed = compressed.reshape(img.shape).astype(np.uint8)
```

```python
plt.subplot(1,2,1)
plt.title("original")
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Compressed')
plt.imshow(compressed)
```
