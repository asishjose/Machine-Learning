```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Load dataset
df = pd.read_csv('Melbourne_housing.csv')

# Step 2: Drop rows with missing values for selected features
df = df[['BuildingArea', 'Price']].dropna()

# Step 3: Use BuildingArea as the feature and Price as target
X = df[['BuildingArea']]
y = df['Price']

# Step 4: Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_linear = lin_reg.predict(X)

# Step 5: Polynomial Regression (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
y_pred_poly = poly_reg.predict(X_poly)

# Step 6: Plotting
plt.figure(figsize=(10,6))
plt.scatter(X, y, color='lightgray', label='Actual Data', alpha=0.5)
plt.plot(X, y_pred_linear, color='blue', label='Linear Regression')
plt.scatter(X, y_pred_poly, color='red', s=10, label='Polynomial Regression (degree=2)', alpha=0.5)
plt.xlabel('Building Area')
plt.ylabel('Price')
plt.legend()
plt.title('Linear vs Polynomial Regression on Melbourne House Prices')
plt.show()

# Step 7: Evaluation
print("Linear Regression:")
print("MSE:", mean_squared_error(y, y_pred_linear))
print("RÂ²:", r2_score(y, y_pred_linear))

print("\nPolynomial Regression:")
print("MSE:", mean_squared_error(y, y_pred_poly))
print("RÂ²:", r2_score(y, y_pred_poly))
```
