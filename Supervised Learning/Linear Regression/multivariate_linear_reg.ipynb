```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

```python
df = pd.read_csv('homeprices_multivariate.csv')
df
```

```python
df.isnull().sum()
```

```python
df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop('price', axis='columns'), df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

#### home with 3000 sqr ft area, 3 bedrooms, 40 year old

# Hiring Model

```python
df_hire = pd.read_csv('hiring.csv')
df_hire
```

```python
df_hire.isnull().sum()
```

```python
from word2number import w2n
```

```python
df_hire.experience = df_hire.experience.fillna('zero')
```

```python
df_hire.experience = df_hire.experience.apply(w2n.word_to_num)
```

```python
median_test_score = df_hire['test_score(out of 10)'].mean()
```

```python
import math
median_test_score = math.floor(median_test_score)
median_test_score
```

```python
df_hire['test_score(out of 10)'] = df_hire['test_score(out of 10)'].fillna(median_test_score)
```

```python
df_hire.experience = df_hire.experience.replace(0, np.nan)
```

```python
df_hire.experience = df_hire.experience.fillna(df_hire.experience.median())
```

```python
hire_reg = linear_model.LinearRegression()
hire_reg.fit(df_hire.drop('salary($)',axis='columns'),df_hire['salary($)'])
```

```python
hire_reg.predict([[2,9,6]])
```

```python

hire_reg.predict([[12,10,10]])
```

```python
hire_reg.intercept_
```
