```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price, color='red', marker='*')
```

```python
new_df = df.drop('price', axis='columns')
new_df
```

```python
price = df.price
price
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']], df.price)
```

```python
model.predict([[3300]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
m = 135.78767123
x = 3300
b = 180616.43835616432
y = m*x+b
y
```

```python
model.predict([[5000]])
```

### Generate CSV file with list of home price predictions

```python
area_df = pd.read_csv('areas.csv')
area_df.shape
```

```python
p = model.predict(area_df)
p
```

```python
area_df['prices'] = p
area_df
```

```python
area_df.to_csv('prediction.csv')
```

```python
rounded_prices = np.round(p,2)
```

```python
rounded_prices
```

```python
area_df['rounded_price'] = rounded_prices
```

```python
area_df
```

```python
area_df
```

```python
# Step 1: Create the 'formatted_price' column
area_df['formatted_price'] = area_df['rounded_price'].apply(lambda x: f"{x:,.2f}")

# Step 2: Now you can access or print it
print(area_df[['area', 'rounded_price', 'formatted_price']])
```
