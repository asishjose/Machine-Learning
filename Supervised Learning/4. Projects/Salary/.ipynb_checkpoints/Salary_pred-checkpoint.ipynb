#### work_year Feature Treatment

The work_year column represents the year in which the salary record was reported (e.g., 2020, 2021, 2022, etc.). Although it is stored as numeric values, it does not represent a continuous numerical feature like age or salary. Instead, it is a time-based categorical feature where each year corresponds to a distinct category.

Keeping it as a numerical variable would incorrectly imply a linear relationship between years (e.g., that 2025 is "5 units larger" than 2020). To avoid this, we convert work_year into a categorical feature and apply encoding (such as one-hot encoding), which allows the model to learn year-specific salary patterns without assuming continuity.

```python
df['work_year'] = df['work_year'].astype('object')
```

#### remote_ratio Feature Treatment

The remote_ratio column indicates the proportion of remote work in a job (0 = on-site, 50 = hybrid, 100 = fully remote). Although it is stored as numerical values, it represents discrete categories, not a continuous numerical variable.

To ensure the model captures the categorical nature of work setups, we convert remote_ratio into a categorical feature and apply encoding. This allows the model to distinguish between on-site, hybrid, and remote work arrangements correctly.

```python
df['remote_ratio'] = df['remote_ratio'].astype('object')
```

#### 'salary' and 'salary_in_usd' Feature Treatment

The dataset contains both salary (reported in local currencies) and salary_in_usd (converted into a common USD scale). Since our prediction task is to model salary_in_usd, we treat it as the target variable.

The raw salary column introduces redundancy and inconsistency due to different local currencies, so we drop it from the feature set. This ensures the model is trained on consistent and comparable salary data.

```python
df = df.drop(columns=['salary'])
```

The job_title column contains 317 unique job titles, many of which appear only once or a few times. Using it directly in a machine learning model would create too many features if one-hot encoded, leading to sparse data and slower training.

To address this, we group similar job titles into broader categories based on job function or seniority.

This reduces 317 titles → 6–8 meaningful categories.

It retains the predictive power of job roles while keeping the model manageable.

## Preprocessing

#### Ordinal mappings

```python
experience_map = {'EN': 0, 'MI': 1, 'SE': 2, 'EX': 3}
company_size_map = {'S': 0, 'M': 1, 'L': 2}

categorical_features['experience_level'] = categorical_features['experience_level'].map(experience_map)
categorical_features['company_size'] = categorical_features['company_size'].map(company_size_map)
```

```python
num_features = ['work_year', 'experience_level', 'remote_ratio', 'company_size']
cat_features = ['employment_type', 'job_group', 'location']
```

#### PIPELINE

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression

numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)

preprocessor = ColumnTransformer(
    transformers=[
        ('num',numeric_transformer, num_features),
        ('cat',categorical_transformer, cat_features)
    ]
)

model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
```

## Splitting

```python
X = categorical_features
y = df['salary_in_usd']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print('RMSE:', rmse)
print('R2_Score:', r2)
```

```python
df['salary_in_usd'].mean()
```

### Ridge

```python
from sklearn.linear_model import Ridge

ridge_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', Ridge())
])


ridge_model.fit(X_train, y_train)

y_pred = ridge_model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print('RMSE:', rmse)
print('R2_Score:', r2)
```

### tuning

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform

param_dist = {
    'regressor__alpha':uniform(0.001,10)
}

ridge_cv = RandomizedSearchCV(
    estimator=ridge_model,
    param_distributions=param_dist,
    n_iter = 50,
    cv = 5,
    scoring= 'r2',
    random_state=42,
    n_jobs = -1
)

ridge_cv.fit(X_train, y_train)
```

```python
print("Best alpha:", ridge_cv.best_params_)
print("Best CV R²:", ridge_cv.best_score_)
```

```python
y_pred = ridge_cv.predict(X_test)


rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("R² Score:", r2)
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

tree_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', DecisionTreeRegressor(random_state=42))
])

tree_model.fit(X_train, y_train)
```
