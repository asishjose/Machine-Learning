The data set has slightly imbalanced but its fine we can use it.

Balanced Dataset :- Each class has same propotion of distribution
Slightly Imbalanced Dataset:- class distribution is roughly 60-40, 70-30, or even 80-20, depending on the problem.
Imbalanced Dataset:- When the class distribution is highly skewed, such as 90-10, 95-5, or even more extreme imbalances.

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
train1 = pd.get_dummies(train,drop_first= True).astype('int')
train1.head()
```

```python
X = train1.drop('Survived',axis = 1)
y = train1['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.8,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Model1 = LinearRegression()
Model1.fit(X_train,y_train)
```

```python
y_train_pred = Model1.predict(X_train)
y_test_pred = Model1.predict(X_test)
```

```python
from sklearn.metrics import r2_score
display(r2_score(y_train,y_train_pred))
display(r2_score(y_test,y_test_pred))
```

```python
Model2 = LogisticRegression()
Model2.fit(X_train,y_train)
```

```python
y_train_pred2 = Model2.predict(X_train)
y_test_pred2 = Model2.predict(X_test)
```

```python
confusion_matrix(y_train,y_train_pred2)
```

```python
confusion_matrix(y_test,y_test_pred2)
```

Support :- The number of samples each metrics was calculated on.

Macro Average:- The Average of Precision, recall and F1 Score between classes. Macro avg doesn’t class imbalance into effort, so if you do have class imbalances, you must focus on Macro Avg

Weighted Average:- The Weighted average of precision, recall and F1 Score between classes. Weighted means each metric is calculated with respect to how many samples there are in each class. This metric will favor the majority class.
