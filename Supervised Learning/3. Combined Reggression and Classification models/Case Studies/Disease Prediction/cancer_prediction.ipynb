```python
df = df.drop(columns=['id', 'Unnamed: 32'])
```

```python
df.isnull().sum()
```

#### Splitting

```python
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df.head()
```

### Train_Test_Spit

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df,y, test_size=0.2, random_state=5)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### Primary Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

report = classification_report(y_test, y_pred)
print(report)

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

### Handling class Imbalance

#### SMOTE

```python
import imblearn
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X_scaled_df,y)

y_sm.value_counts()
```

### Cross Validation

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

skfold = StratifiedKFold(n_splits=5)
model = LogisticRegression()
scores = cross_val_score(model, X_scaled_df, y, cv=skfold)

print('Cross Validation Scores:', np.mean(scores))
```

### Model Training

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sm,y_sm, test_size=0.2, random_state=5)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model = LogisticRegression()
model.fit(X_sm, y_sm)
```

```python
y_pred = model.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

report = classification_report(y_test, y_pred)
print(report)

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
y_encoded = y.map({'B': 0, 'M': 1})
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df,y_encoded, test_size=0.2, random_state=5)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform

model = LogisticRegression()

param_dist = {
    'C': uniform(loc=0.01, scale=10),
    'penalty':['l1','l2'],
    'solver':['liblinear'],
    'max_iter':[100,200,300]
}

random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_dist,
    n_iter=20,
    cv=5,
    scoring='accuracy',
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)
```

```python
random_search.best_params_
```

```python
best_model = random_search.best_estimator_
best_model
```

```python
y_pred = best_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

report = classification_report(y_test, y_pred)
print(report)

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

### ROC-AUC

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
```

```python
y_probs = best_model.predict_proba(X_test)[:, 1]
y_test = y_test
```

```python
roc_auc = roc_auc_score(y_test, y_probs)
roc_auc
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], 'k--')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
```

##### What the plot shows:
The blue ROC curve is very close to the top-left corner of the plot.

The diagonal dashed line represents a random classifier (ROC-AUC = 0.5).

Your curve hugs the top-left, indicating very high True Positive Rate with very low False Positive Rate.

Interpretation:
Your model has excellent discrimination ability between positive and negative classes.

The ROC-AUC value is very close to 1 (probably above 0.95), which means your model almost perfectly separates the classes.

This suggests your classifier is highly effective on this dataset.

Summary:
ROC curve near top-left corner → very good model.

Area Under Curve (AUC) close to 1 → high predictive power.
