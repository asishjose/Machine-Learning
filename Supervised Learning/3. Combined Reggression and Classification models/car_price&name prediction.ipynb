```python
doors_map = {'two':2, 'four':4}
cylinders_map = {'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'eight': 8, 'twelve': 12}

df['doornumber'] = df['doornumber'].map(doors_map)
df['cylindernumber'] = df['cylindernumber'].map(cylinders_map)
df[['doornumber','cylindernumber']]
```

## One-hot encoding

```python
nominal = df[['CompanyName',
 'fueltype',
 'aspiration',
 'carbody',
 'drivewheel',
 'enginelocation',
 'enginetype',
 'fuelsystem']]
nominal
```

```python
nominal = pd.get_dummies(nominal,drop_first=True).astype(int)
nominal
```

Correlation-based methods: Correlation can be misleading on sparse one-hot features.

PCA / Dimensionality reduction: Can work but makes interpretability harder and may mix categories in combinations, losing clear meaning.

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.01)
X_reduced = selector.fit_transform(nominal)
```
