The dataset is called Electric Vehicle Population Data (Washington State, USA).
Shape: (250,659 rows × 17 columns)

Each row = one registered EV in Washington.

**Columns:**

- VIN (1-10) → First 10 characters of VIN (unique vehicle identifier).

Not useful for ML (like an ID).

- County → County in Washington State.

Categorical (225 unique values, King most common).

- City → City name.

Categorical (822 unique values, Seattle most common).

- State → State code (almost all WA).

Low variance → can probably drop.

- Postal Code → ZIP code.

High cardinality, could be useful if aggregated (e.g., clustering EV adoption).

- Model Year → Year vehicle was manufactured.

Numeric (2000 → 2026).

- Make → Manufacturer (e.g., Tesla, Chevrolet).

Categorical (46 unique, Tesla dominates with 42%).

- Model → Specific car model (e.g., Model Y, Leaf).

Categorical (178 unique).

- Electric Vehicle Type → Target candidate: BEV (80%) vs PHEV (20%).

Binary classification.

- Clean Alternative Fuel Vehicle (CAFV) Eligibility → Whether vehicle qualifies for CAFV program.

3 categories: Eligible, Not Eligible, Unknown.

- Electric Range → Vehicle’s electric-only range (miles).

Numeric, very relevant for BEV vs PHEV classification.

- Base MSRP → Manufacturer’s Suggested Retail Price (price of the car).

Numeric (though often missing or skewed).

- Legislative District → District number.

Numeric (categorical-like, depends on district boundaries).

- DOL Vehicle ID → Unique ID.

Not useful (drop).

- Vehicle Location → Geo coordinates (latitude, longitude).

Could be engineered into clusters / regions.

- Electric Utility → Utility provider.

Categorical (many categories).

- 2020 Census Tract → Census data identifier.

High cardinality → might not be very useful unless grouped.
