```python
import pandas as pd
from sklearn import model_selection
from sklearn.ensemble import AdaBoostClassifier
```

```python
df = pd.read_csv('diabetes.csv')
array = df.values
X = array[:,0:8]
y = array[:,8]
seed = 7
num_trees = 30
```

```python
kfold = model_selection.KFold(n_splits=10, random_state=seed, shuffle=True)
model = AdaBoostClassifier(n_estimators=num_trees, random_state=seed)
results = model_selection.cross_val_score(model, X, y, cv=kfold)
print(results.mean())
```

```python
from sklearn import svm
from xgboost import XGBClassifier
clf = XGBClassifier()

seed = 7
num_trees = 30
kfold = model_selection.KFold(n_splits=10, random_state=seed, shuffle=True)
model = XGBClassifier(n_estimators=num_trees, random_state=seed)
results = model_selection.cross_val_score(model, X,y,cv=kfold)
print(results.mean())
```

### Cross VAlidation

```python
from sklearn.datasets import load_iris
df = load_iris()
df
```

```python
data_input = df.data
data_output = df.target
```

```python
print(data_input)
```

```python
print(data_output)
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True)
```

```python
print("Train set        Test set      ")
for train_set, test_set in kf.split(data_input):
    print(train_set, test_set)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_cls = RandomForestClassifier(n_estimators=10)
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(rf_cls, data_input, data_output, scoring='accuracy', cv=10))
```

```python
accuracy = cross_val_score(rf_cls, data_input, data_output, scoring='accuracy', cv=10).mean()*100
print('Accuracy of Random Forest is:', accuracy)
```
